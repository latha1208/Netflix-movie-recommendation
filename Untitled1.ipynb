```python
# Load the datasets
import pandas as pd

movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')

# Show first few rows of each to verify upload
print(movies.head())
print(ratings.head())
```

```python
# Create a user-movie matrix
user_movie_matrix = ratings.pivot_table(index='userId', columns='movieId', values='rating')

# Replace missing ratings with 0
user_movie_matrix.fillna(0, inplace=True)

# Show the matrix
print(user_movie_matrix.head())
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Calculate similarity between movies (transpose the matrix)
movie_similarity = cosine_similarity(user_movie_matrix.T)

# Convert it into a DataFrame for easier lookup
movie_similarity_df = pd.DataFrame(movie_similarity, index=user_movie_matrix.columns, columns=user_movie_matrix.columns)

# Show a part of the similarity matrix
print(movie_similarity_df.head())
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Calculate similarity between movies (transpose the matrix)
movie_similarity = cosine_similarity(user_movie_matrix.T)

# Convert it into a DataFrame for easier lookup
movie_similarity_df = pd.DataFrame(movie_similarity, index=user_movie_matrix.columns, columns=user_movie_matrix.columns)

# Show a part of the similarity matrix
print(movie_similarity_df.head())
```

```python
# Function to recommend movies
def recommend_movies(movie_id, num_recommendations=5):
    similar_scores = movie_similarity_df[movie_id].sort_values(ascending=False)
    similar_movie_ids = similar_scores.iloc[1:num_recommendations+1].index
    recommended_movies = movies[movies['movieId'].isin(similar_movie_ids)]
    return recommended_movies[['title']]

# Example: Recommend movies similar to Toy Story (movieId = 1)
recommend_movies(1, 5)
```

```python
# Function to search for movieId by movie title
def find_movie_id(movie_title):
    result = movies[movies['title'].str.contains(movie_title, case=False, na=False)]
    return result[['movieId', 'title']]
```

```python
# Find movieId for Titanic
find_movie_id('Titanic')

```
